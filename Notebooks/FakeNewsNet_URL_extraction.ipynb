```python
gos_fake_df = pd.read_csv("/content/gdrive/My Drive/Group_Project_530/FakeNews_Dataset/gossipcop_fake.csv")
gos_real_df = pd.read_csv("/content/gdrive/My Drive/Group_Project_530/FakeNews_Dataset/gossipcop_real.csv")
pol_fake_df = pd.read_csv("/content/gdrive/My Drive/Group_Project_530/FakeNews_Dataset/politifact_fake.csv")
pol_real_df = pd.read_csv("/content/gdrive/My Drive/Group_Project_530/FakeNews_Dataset/politifact_real.csv")
```

```python
dataframes = [gos_fake_df,pol_fake_df, gos_real_df, pol_real_df]
for ind,df in enumerate(dataframes):
  if ind<2:
    df['Label'] = 1
  else:
    df['Label'] = 0
  df.drop(['id'], axis=1, inplace=True)
```

```python
all_df = pd.concat(dataframes)
all_df = all_df.sample(n=len(all_df), random_state=42)
all_df = all_df.reset_index().drop(['index'], axis=1)
all_df.drop(['title', 'tweet_ids'], axis=1, inplace=True)
all_df.head()
```

```python
def extractBody(url):

  downloaded = trafilatura.fetch_url(url) 

  if downloaded is None:
    return 'N/A'
    
  else:
    extracted= trafilatura.extract(downloaded)
    return extracted
```

```python
all_df['Body'] = all_df.news_url.apply(lambda x: extractBody(x))
```

```python
all_df.drop(['news_url'], axis=1, inplace=True)
```
